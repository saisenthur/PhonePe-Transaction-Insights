```python
import os
import json
import pandas as pd

# Set your base path
base_path = "C:/Users/saisenthur/OneDrive/Desktop/Courses/Guvi/Data Science 2/projects/2_PhonePe Transaction Insights/Data_new/data"

# Helper to get quarter number from filename
def get_quarter(filename):
    return int(filename.strip(".json"))

# 1. Aggregated Transaction
def extract_aggregated_transaction():
    path = f"{base_path}/aggregated/transaction/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Count': [], 'Amount': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    trans_data = data.get('data', {}).get('transactionData')
                    if trans_data:
                        for record in trans_data:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['Transaction_type'].append(record['name'])
                            clm['Count'].append(record['paymentInstruments'][0]['count'])
                            clm['Amount'].append(record['paymentInstruments'][0]['amount'])

    Agg_Trans = pd.DataFrame(clm)
    Agg_Trans.to_csv("Agg_Transaction.csv", index=False)

# 2. Aggregated User
def extract_aggregated_user():
    path = f"{base_path}/aggregated/user/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'Brand': [], 'Count': [], 'Percentage': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    users_data = data.get('data', {}).get('usersByDevice')
                    if users_data:
                        for record in users_data:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['Brand'].append(record['brand'])
                            clm['Count'].append(record['count'])
                            clm['Percentage'].append(record['percentage'])

    Agg_User = pd.DataFrame(clm)
    Agg_User.to_csv("Agg_User.csv", index=False)

# 3. Aggregated Insurance
def extract_aggregated_insurance():
    path = f"{base_path}/aggregated/insurance/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'Count': [], 'Amount': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    insurance_data = data.get('data', {}).get('transactionData')
                    if insurance_data:
                        for record in insurance_data:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['Count'].append(record['paymentInstruments'][0]['count'])
                            clm['Amount'].append(record['paymentInstruments'][0]['amount'])

    Agg_Insurance = pd.DataFrame(clm)
    Agg_Insurance.to_csv("Agg_Insurance.csv", index=False)

# 4. Map Transaction
def extract_map_transaction():
    path = f"{base_path}/map/transaction/hover/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [], 'Amount': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    trans_data = data.get('data', {}).get('hoverDataList')
                    if trans_data:
                        for record in trans_data:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['District'].append(record['name'])
                            clm['Count'].append(record['metric'][0]['count'])
                            clm['Amount'].append(record['metric'][0]['amount'])

    Map_Trans = pd.DataFrame(clm)
    Map_Trans.to_csv("Map_Transaction.csv", index=False)

# 5. Map User
def extract_map_user():
    path = f"{base_path}/map/user/hover/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'RegisteredUsers': [], 'AppOpens': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    hover_data = data.get('data', {}).get('hoverData')
                    if hover_data:
                        for district, metrics in hover_data.items():
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['District'].append(district)
                            clm['RegisteredUsers'].append(metrics.get('registeredUsers', 0))
                            clm['AppOpens'].append(metrics.get('appOpens', 0))

    Map_User = pd.DataFrame(clm) 
    Map_User.to_csv("Map_User.csv", index=False)

# 6. Map Insurance
def extract_map_insurance():
    path = f"{base_path}/map/insurance/hover/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [], 'Amount': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    ins_data = data.get('data', {}).get('hoverDataList')
                    if ins_data:
                        for record in ins_data:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['District'].append(record['name'])
                            clm['Count'].append(record['metric'][0]['count'])
                            clm['Amount'].append(record['metric'][0]['amount'])

    Map_Insurance = pd.DataFrame(clm)
    Map_Insurance.to_csv("Map_Insurance.csv", index=False)

# 7. Top Transaction
def extract_top_transaction():
    path = f"{base_path}/top/transaction/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [], 'Amount': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    entities = data.get('data', {}).get('districts')
                    if entities:
                        for record in entities:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['District'].append(record['entityName'])
                            clm['Count'].append(record['metric'].get('count', 0))
                            clm['Amount'].append(record['metric'].get('amount', 0))

    Top_Trans = pd.DataFrame(clm)
    Top_Trans.to_csv("Top_Transaction.csv", index=False)

# 8. Top User
def extract_top_user():
    path = f"{base_path}/top/user/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'RegisteredUsers': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    entities = data.get('data', {}).get('districts')
                    if entities:
                        for record in entities:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['District'].append(record['name'])
                            clm['RegisteredUsers'].append(record['registeredUsers'])

    Top_User = pd.DataFrame(clm)
    Top_User.to_csv("Top_User.csv", index=False)

# 9. Top Insurance
def extract_top_insurance():
    path = f"{base_path}/top/insurance/country/india/state/"
    clm = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [], 'Amount': []}

    for state in os.listdir(path):
        for year in os.listdir(f"{path}{state}/"):
            for file in os.listdir(f"{path}{state}/{year}/"):
                with open(f"{path}{state}/{year}/{file}", 'r') as f:
                    data = json.load(f)
                    entities = data.get('data', {}).get('districts')
                    if entities:
                        for record in entities:
                            clm['State'].append(state)
                            clm['Year'].append(year)
                            clm['Quarter'].append(get_quarter(file))
                            clm['District'].append(record['entityName'])
                            clm['Count'].append(record['metric'].get('count', 0))  # Default to 0 if 'count' is missing
                            clm['Amount'].append(record['metric'].get('amount', 0))  # Default to 0 if 'amount' is missing

    Top_Insurance = pd.DataFrame(clm)
    Top_Insurance.to_csv("Top_Insurance.csv", index=False)

# Execute all
extract_aggregated_transaction()
extract_aggregated_user()
extract_aggregated_insurance()
extract_map_transaction()
extract_map_user()
extract_map_insurance()
extract_top_transaction()
extract_top_user()
extract_top_insurance()
```

```python
import os
import shutil

# Define the current folder where the CSV files are located (adjust path if needed)
current_folder = "C:/Users/saisenthur/OneDrive/Desktop/Courses/Guvi/Data Science 2/projects/2_PhonePe Transaction Insights/myenv/Scripts"

# Define the new folder where you want to move the CSV files
new_folder = "C:/Users/saisenthur/OneDrive/Desktop/Courses/Guvi/Data Science 2/projects/2_PhonePe Transaction Insights/CSV files/"

# Create the new folder if it doesn't exist
if not os.path.exists(new_folder):
    os.makedirs(new_folder)

# List of CSV file names you want to move (modify if needed)
csv_files = [
    "agg_transaction.csv",
    "agg_user.csv",
    "agg_insurance.csv",
    "map_transaction.csv",
    "map_user.csv",
    "map_insurance.csv",
    "top_transaction.csv",
    "top_user.csv",
    "top_insurance.csv"
]

# Move each CSV file to the new folder
for file in csv_files:
    source_path = os.path.join(current_folder, file)
    destination_path = os.path.join(new_folder, file)
    
    # Check if the file exists in the current folder before moving
    if os.path.exists(source_path):
        shutil.move(source_path, destination_path)
        print(f"Moved {file} to {new_folder}")
    else:
        print(f"File {file} not found in {current_folder}")
```
